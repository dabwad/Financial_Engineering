```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

ticker = 'AAPL'
start_date='2022-08-01'
df=yf.download(ticker, start=start_date)

print(df.head())
```

```python
short_window = 5 # 단기
long_window = 50 # 장기
data = df.copy()

#지수가중이동평균선 계산
data['Short_MA'] = data['Close', 'AAPL'].ewm(span=short_window, adjust=False).mean()
data['Long_MA'] = data['Close', 'AAPL'].ewm(span=long_window, adjust=False).mean()
data=data[long_window:].copy() #초기데이터 제거

cash_init = 1000000
cash = cash_init
asset=[cash]
pos = 0

for i in range(1, len(data)):
    if pos == 0:
        if data['Short_MA'].iloc[i] > data['Long_MA'].iloc[i] and \
        data['Short_MA'].iloc[i-1] < data['Long_MA'].iloc[i-1]:
            pos = 1
            entry_price=data['Close', 'AAPL'].iloc[i]
            num=int(cash/entry_price)
            cash-=entry_price*num
    elif pos==1:
        if data['Short_MA'].iloc[i] < data['Long_MA'].iloc[i] and \
        data['Short_MA'].iloc[i-1] > data['Long_MA'].iloc[i-1]:
            pos = 0
            entry_price=data['Close', 'AAPL'].iloc[i]
            cash+=entry_price*num

    # 자산 가치 갱신 기록
    if pos == 0:
        asset.append(cash)
    elif pos == 1:
        asset.append(cash+data['Close', 'AAPL'].iloc[i]*num)

# 누적 수익률 계산 출력
data['Cumulative_Return']=np.array(asset)/cash_init
final_cum_return=data['Cumulative_Return'].iloc[-1]-1
print(f'Final cumulative return of the strategy: {100*final_cum_return: .2f}%')
```

```python
short_window = 5 # 단기
long_window = 20 # 장기
data = df.copy()

#지수가중이동평균선 계산
data['Short_MA'] = data['Close', 'AAPL'].ewm(span=short_window, adjust=False).mean()
data['Long_MA'] = data['Close', 'AAPL'].ewm(span=long_window, adjust=False).mean()
data=data[long_window:].copy() #초기데이터 제거


#포지션과 시그널 계산
data['Position']=np.where(data['Short_MA']>data['Long_MA'], 1, 0)
data['Signal']=data['Position'].diff().fillna(0) #1: 매수, -1: 매도

prices=data['Close', 'AAPL'].values
signals=data['Signal'].values

cash_init = 1000000
cash = cash_init
asset=np.zeros(len(data))
asset[0]=cash
pos = 0

for i in range(1, len(data)):
    if pos == 0:
        if signals[i]==1:
            pos = 1
            entry_price=prices[i]
            num=int(cash/entry_price)
            cash-=entry_price*num
    elif pos==1:
        if signals[i]==-1:
            pos = 0
            entry_price=prices[i]
            cash+=entry_price*num

    # 자산 가치 갱신 기록
    if pos == 0:
        asset[i]=cash
    elif pos == 1:
        asset[i]=cash+prices[i]*num

# 누적 수익률 계산 출력
data['Cumulative_Return']=np.array(asset)/cash_init
final_cum_return=data['Cumulative_Return'].iloc[-1]-1
print(f'Final cumulative return of the strategy: {100*final_cum_return: .2f}%')

#시각화를 위한 매수, 매도 시점의 가격 칼럼 계산
data['Buy_Price'] = np.where(signals==1, prices, np.nan)
data['Sell_Price'] = np.where(signals==-1, prices, np.nan)
```

```python
fig, ax = plt.subplots(2, 1, sharex=True, height_ratios=(8, 2), figsize=(10,8))

data['Close'].plot(ax=ax[0], label='Close')
data['Short_MA'].plot(ax=ax[0], label='Short MA', linewidth=1)
data['Long_MA'].plot(ax=ax[0], label='Long MA', linewidth=1)
data['Buy_Price'].plot(ax=ax[0], label='Buy', marker='^', color='b', markersize=8)
data['Sell_Price'].plot(ax=ax[0], label='Sell', marker='v', color='r', markersize=8)

ax[0].set_title(f'{ticker} Moving Average Crossover Trades', fontsize=18)
ax[0].set_ylabel('Price($)', fontsize=12)
ax[0].legend(fontsize=12)
ax[0].grid(alpha=0.3)

data['Position'].plot(ax=ax[1])
ax[1].set_xlabel('Date', fontsize=12)
ax[1].set_ylabel('Position', fontsize=12)
ax[0].grid(alpha=0.3)

plt.xticks(rotation=0)
plt.tight_layout()
plt.show

plt.savefig('chart.png', dpi=300, bbox_inches='tight')
```
